```python
# Create s function using locals()


def demonstrate_locals():
  """
  Demonstrates the locals() dictionary and its behavior
  locals() returns a dictionary of current local symbol table
  """


  # Create several local variables
  name = "Alice"
  age = 30
  scores = [85, 92, 78]
  is_active = True

  print("=== LOCAL VARIABLES CREATED ===")
  print(f"name:{name}")
  print(f"age:{age}")
  print(f"scores: {scores}")
  print(f"is_active: {is_active}")


  print("\n === locals() DICTIONARY ===")
  # print the locals() dictionary
  local_vars = locals()
  for key, value in local_vars.items():
    if not key.startswith('_'): #Skip internal variables
      print(f"{key}: {value}")


  print("\n=== ATTEMPTING TO MODIFY VIA locals() ===")
  # Try to modify a local variable through locals()
  print("Before modification - age:", age)


  # Attempt to modify via locals() dictionary
  local_vars['age'] = 35
  print("After locals() modification - age:", age) # Won't change
  print("locals()['age]:", local_vars['age']) # This changed


  # The actual local variable 'age'  remains unchanged
  # locals() returns a COPY in some contexts, not a refrence



  print("\n=== CREATING NEW LOCAL VIA locals() ====")
  # Try to create new local variable via locals()
  local_vars['new_variable'] = "I am new"
  print("new_variable" in locals()) # Usually False
  # print(new_variable) $ This would cause NameError!



def show_locals_behavior():
  """
  Shows different behavior of locals() in various contexts
  """

  x = 10
  y = 20

  print("\n=== locals() IN FUNCTION ===")
  print("locals() keys:", [k for k in locals().keys() if not k.startswith('_')])



  # locals() in function scope typically returns a copy
  local_copy = locals()
  local_copy['x'] = 999
  print("After locals() modification - x", x)
  print("locals()['x]:", local_copy['x'])




# Run demonstrations
demonstrate_locals()
show_locals_behavior()

# Compare with globals() behavior
print("\n=== COMPARISON: globals() vs locals() ===")
global_test = "global_value"

def compare_scope_dicts():
    local_test = "local_value"

    print("globals() can modify:", 'global_test' in globals())
    print("locals() can modify:", 'local_test' in locals())

    # This works - globals() returns reference
    globals()['global_test'] = "modified_global"

    # This doesn't work the same way - locals() often returns copy
    locals()['local_test'] = "modified_local"
    print("Local test after modification:", local_test)  # Still "local_value"

compare_scope_dicts()
print("Global test after modification:", global_test)  # "modified_global"



```

```python
# Demonstrate LEGB rule without nesting


global_message = "I am a global variable"
GLOBAL_CONSTANT = 100


def demonstrate_legb():
  """
  Demonstrates LEGB rule: Local, Enclosing, Global, Built-in
  Shows Local, Global and Built-in scopes without nesting
  """

  # LOCAL SCOPE - This is the 'L' in LEGB
  local_message = "I am the local variable"
  local_count = 5


  print("=== LOCAL SCOPE (L) ===")
  print(f"Local variable: {local_message}")
  print(f"Local count: {local_count}")


  print("\n=== GLOBAL SCOPE (G) ===")
  # Accessing global variables - Python finds them after checking local scope

  print(f"Global variable: {global_messsage}")
  print(f"Global constant: {GLOBAL_CONSTANT}")


  print("\n=== BUILT-IN SCOPE (B) ==")
  # Accessing built-in functions - Python finds tham after checking the Local and Global


  sample_list = [1,2,3,4,5]


  # Built-in function 'len' - from Python's built-in namespace

  length = len(sample_list)
  print(f"Built-in len() result: {length}")



  # More built-in functions
  max_value = max(sample_list)
  min_value = min(sample_list)
  sum_value = sum(sample_list)



  print(f"Built-in max(): {max_value}")
  print(f"Built-in min(): {min_value}")
  print(f"Built_in sum(): {sum_value}")


  # Built-in constants
  print(f"Built-in True: {True}")
  print(f"Built-in None: {None}")



def show_name_resolution():
  """
  Shows how Python resolves names using LEGB rule
  """

  # This will use the local version when available

  len = "I am shadowing built-in len " # Not to do in real code

  print("\n=== NAME RESOLUTION ORDER ===")
  print(f"Local 'len' shadows built-in {len}")


  # To shadow the builf-in len after shadowing, we nee to be explicit
  import builtins
  real_len = builtins.len([1,2,3])
  print(f"Real built-in len: {real_len}")




def demonstrate_scope_conflict():
  """
  Shows what happens when names conflict across scopes
  """

  # This has same name as global - creates LOCAL variable (shadowing)

  global_message = "I am local, shadowing global"


  print("\n=== SCOPE CONFLICT ===")
  print(f"local wins: {global_message}")    # uses local version

  # To access the global one expliitly
  import builtins
  print(f"Global versio: {globals()['global_message']}")



# Demonstrate LEGB rule

demonstrate_legb()

# show name resolution in action
show_name_resolution()

demonstrate_scope_conflict()

# prove global remains unchanged
print(f"\n=== GLOBAL STILL INTACT ===")
print(f"Original global: {global_message}")
```

```python
# Global mutable object

global_list = [1,2,3]
global_counter = 100


def modify_without_global():
  """
  Modifying mutable objects works without 'global' keyword
  Because we are modifying the object, not rebinding the varibale
  """

  print("=== MODIFYING THE MUTABLE OBJECT ===")
  print(f"Before: {global_list}")


  # This works - modifying the object itself
  global_list.append(4) # modifies existing list
  global_list[0] = 999     # modifies existing list


  print(f"After: {global_list}")



def reassign_needs_global():
  """
  Reassignment needs 'global' keyword
  Because we are creating a new object and rebinding the variable
  """

  print("\n=== REASSIGNMENT (WILL FAIL) ===")
  print(f"Before: {global_counter}")


  try:
    # global_counter += 1
    print(f"After: {global_counter}")
  except UnboundLocalError as e:
    print(f"Error: {e}")



def reassign_with_global():
  """
  Reassignment works with 'global' keyword
  """

  print("\n=== REASSIGNMENT WITH 'global' ===")
  global global_counter # Declare we are using the global variable



  print(f"Before: {global_counter}")
  global_counter += 1   # Now works - rebinding global variable
  print(f"After: {global_counter}")



# Demonstrate the behaviors
modify_without_global()
reassign_needs_global()
reassign_with_global()

# Show final state
print(f"\n=== FINAL STATE ===")
print(f"global_list: {global_list}")
print(f"global_counter: {global_counter}")
```

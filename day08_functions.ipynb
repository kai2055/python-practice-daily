```python
# Create a function that takes 3 numbers and returns the middle value
# Input: (5,2,8) -> 5

def mid_value(a, b, c):
  lst = [a,b,c]
  lst.sort()
  return(lst[1])


print(mid_value(1,5,2))
```

```python
# Create a function that return a temperature in fahrenheit, kelvin when given in celsius

def temperature_converter(temp):
  fahrenheit = temp * 1.8 + 32
  kelvin = temp + 273.15

  return fahrenheit, kelvin


f, k = temperature_converter(100)
print(f,k)
```

```python
# Create a function that adds the sum of all parameters as a new element at the end

def end_element_adder(lst):
  lst.append(sum(lst))
  return lst


print(end_element_adder([1,2,3,4,5]))



# Alternative more explicit version
def end_element_adder_explicit(lst):
  # Create a new list with all elements from original
  new_list = list(lst)
  # Calculate and append sum
  new_list.append(sum(new_list))
  return new_list

print(end_element_adder_explicit([1,2,3,4,5]))

```

```python
# Create a function that takes a string and returns a dict:
# {'Original': string, 'uppercase': upper, 'lowercase':lower, length:len}

def string_dict_returner(word):
  my_dict = {
      'Original': word,
      'Uppercase': word.upper(),
      'Lowercase': word.lower(),
      'Length': len(word)}

  return my_dict


returned_dict = string_dict_returner('samuRai')
print(returned_dict)
```

```python
# Create a function for calculating rectangle properties

def rectangle_property(length, width, unit="meters"):
  return length * width, 2 * (length + width), unit


area, perimeter, unit = rectangle_property(2,4, "cm")
print(f"Perimeter: {perimeter}\nArea: {area}\nunit: {unit}")
```

```python
# Create a function that calculates statistics

def statistics_returner(*numbers):
  lst = numbers
  return "Empty" if numbers == () else {
      "count": len(lst),
      "sum": sum(lst),
      "mean": sum(lst)/len(lst),
      "min": min(lst),
      "max": max(lst),
  }


print(statistics_returner())
```

```python
# Create a function that filters arguments:
# 'positive' 'negative' , 'even', 'odd'

def is_positive(n):
  if n > 0:
    return n

def is_negative(n):
  if n < 0:
    return n

def is_zero(n):
  if n == 0:
    return n

def number_filter(*args):
  result = filter(is_positive,
              filter(is_negative,
                filter(is_zero, args)))
  return result


print(number_filter(1,2,3,4,56,7,"e", "f"))
```

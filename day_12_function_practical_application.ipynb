```python
# Create a dictionary merger

previous_profile = {
    'name': 'Nikhil',
    'age': 26,
    'nationality': 'Nepalese',
    'address': 'kathmandu'
}


updated_profile = {
    'address': 'berlin',
    'job': 'working student',
    'age': 27,
    'company': 'Mercedes'
}



def dictionary_merger(*dicts):

  # Improved: Handle any number of dictionaries
  if len(dicts) == 0:
    return {}

  # start with the first dictionary as base
  complete_profile = dicts[0].copy() # IMPROVED: Use copu to avoid modifying original

  # Process all remaining dictionaries
  for dict_index in range(1, len(dicts)):
    current_dict = dicts[dict_index]

    for key, value in current_dict.items():
      # ORIGINAL CRITICAL BUG: Using dict2.values() instead of actual value

      # complete_profile[keys] = dict2.values() # WRONG: This assigns ALL values

      # CORRECTED: Use the actual value for the current key
      if key in complete_profile:
        # ORIGINAL ISSUE: No strategy for handling conflicts
        # was just overwriting without consideration

        # IMPROVED: Explicit conflict resolution (later dict wins)

        complete_profile[key] = value # Update with new value
        print(f"Updated key {key}: {complete_profile[key]}") # Debug info

      else:
        complete_profile[key] = value # Add new key-value pair
        print(f"Added new key {key}: {value}")

    return complete_profile




result = dictionary_merger(previous_profile, updated_profile)
print("\n=== FINAL RESULT ===")
print(result)


```

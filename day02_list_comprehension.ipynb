```python
# Create  a nested list comprehension for multiplication table (1-5)

# First solving the above problem through loops

multiplication_list = [[0 for i in range(5)] for j in range(5)]

for a in range(1,6):
  for b in range(1,6):
    multiplication_list[a - 1] [b - 1] = a * b


print(multiplication_list)



# Using list comprehension

multiplication_table = [[a * b for b in range(1,6)] for a in range(1,6)]
print(multiplication_table)
```

```python
# Extract all numbers from a mixed list

mixed = [1, "hello", 3.14, "world", 42, None, True, [1,2]]


number_extractor = [ x for x in mixed if (type(x) == int) or (type(x) == float)]
print(number_extractor)
```

```python
# Create Cartesian product of two lists

list1 = [1, 2, 3]
list2 = ['a', 'b']

cartesian_product = [(list1[i], list2[j]) for i in range(len(list1)) for j in range(len(list2))]
print(cartesian_product)

# Another approach

cart_prod = [(x,y) for x in list1 for y in list2]
print(cart_prod)

```

```python
# Remove  dupliates while preserving order (without using set)

nums = [1, 2, 3, 2, 4, 1, 5, 3]

filtered_num = [n for n in nums if nums.count(n) == 1]
print(filtered_num)

# The above removes element that has been repeated more than once
# So only the numbers that are not repeated are in the list

seen = set()
unique_nums = [x for x in nums if x not in seen and not seen.add(x)]
print(unique_nums)

```

```python
# Create a list of word lengths, but onoy for words linger than 3 chars

phrase = " I love to code in Python every day"
words_length = []
# Solving it with loop first

split_phrase = phrase.split()
print(split_phrase)
for words in split_phrase:
  if len(words) > 3:
    words_length.append(len(words))

print(words_length)


# List comprehension approach

string_length = [len(words) for words in phrase.split() if len(words) > 3]
print(string_length)
```

```python
# Create a function that takes a string and returns its length as an integer


def string_length(word: str)-> int:
  return len(word)

print(f"{string_length("123")}")
```

```python
# Create a function that takes a list of numbers and returns (min, max, average) as a tuple

list1 = [1,2,3,4,5,6]


def multtiple_operations(lst):

  avg = (sum(lst))/(len(list1))
  m = max(lst)
  mi = min(lst)

  return avg, m, mi


print(f"{multtiple_operations(list1)}")
```

```python
# Create a function that takes a number n and returns a list of all
# even numbers from 0 to n


def even_numbers(n: int)-> list:
  return [x for x in range(0,n+1) if n % 2 == 0]

print(even_numbers(100))
```

```python
# Improved version of the above word counter case

def word_frequency_counter(senten: str)-> dict:
  # Clean the sentence first to handle edge cases
  cleaned_sentence = senten.strip()

  # Word count - more robust
  words = cleaned_sentence.split()
  words_count = len(words)

  # Vowel count - using sum() with generator expression (more Pythonic)
  vowels_count = sum(1 for char in cleaned_sentence.lower() if char in 'aeiou')

  # Character count - excluding spaces if needed
  char_count_no_spaces = len(cleaned_sentence.replace(" ", ""))
  char_count_with_spaces = len(cleaned_sentence)

  return {
      "word_count": words_count,
      "vowel_count": vowels_count,
      "character_count_no_cpaces": char_count_no_spaces,
      "character_count_with_spaces": char_count_with_spaces,
      "average_word_length": char_count_no_spaces / words_count if words_count > 0 else 0

  }


print(word_frequency_counter(sentence))
```

```python
# Most concise version of the above code

def word_frequency_counter(senten: str)-> dict:
  cleaned = senten.strip()
  words = cleaned.split()
  lower_sentence = cleaned.lower()

  return {
      'words': len(words),
      'vowels': sum(lower_sentence.count(vowel) for vowel in 'aeiou'),
      'chars_no_spaces': len(cleaned.replace(' ', '')),
      'chars_with_spaces': len(cleaned)

  }

sentence = "This is the sentence with which is to be given to the function."

print(word_frequency_counter(sentence))
```

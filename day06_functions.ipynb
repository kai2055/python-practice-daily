```python
# Create a function that takes a number:
# - if positive, return the number as is (int/float)
# - if negative, return "Negative" (string)
# - if zero, return None


def conditional_return(num):
  if  num > 0:
    return num
  elif num < 0:
    return "Negative"
  else:
    return 0


print(conditional_return(1))
print(conditional_return(-1))
print(conditional_return(0))
```

```python
# Create a function with a list as default parameter
# Call it multiple times without passing arguments
# Observe the unexpected behavior, then fix it


def func(items=[]):
  "Function with mutable default argument"
  items.append(len(items))
  return items




# Let's see the unexpected behavior
print("First call:")
result1 = func()
print(f"Result: {result1}")


print("\nSecond call: ")
result2 = func()
print(f"Result: {result2}")

print("\nThird call: ")
result3 = func()
print(f"Result: {result3}")


print("\nCall with explicit argument:")
result4 = func([10, 20])
print(f"Result: {result4}")


# Why does the length keep increasing

# Because the same list object persists beterrn function calls! Each time you call
# func() without arguments
# You are working with the exact same list from the previous call
# The list has grown longer since last time
# So len(items) gives a bigger number each time


# FIXED VERSION

def func_fixed(items = None):
  if items is None:
    items = [] # NEW list every time
  items.append(len(len(items)))
  return items









```

```python
# Create two functions:
# - one that modifies a list in-place (sorts it, returns None)
# - one that returns a NEW sorted list without modifying the original



def modified_list(lst):
  lst.append(1)

  return


def unmodified_list(lst):
  sorted_list = sorted(lst)

  return sorted_list



print(modified_list([1,2,3,4,5,-1,-2,-4]))
print(unmodified_list([1,2,3,4,5,-1,-2,-4]))
```

```python



def flexible_function(operation, *args, **options):
  """
  Perform a flexible operation based on the given parameters.


  Parameters:
    operation (str): The operation to perform. Options: "add", "multiply, "concat
    *args: Number or strings to operate on.
    **options: Additional options:
        - round_to (int): Number of decimal places to round for math ops
        - uppercase (bool): convert concatenated result to uppercase.

  Returns:
    The result of the specified operation.

    """



# Handle missing args

  if not args:
    return "Error: No arguments provided"


  # Handle addition of numbers
  if operation == "add":
    try:
      result = sum(args)
      if "round_to" in options:
        result = round(result, options["round_to"])
      return result
    except TypeError:
      return "Error: 'add' operation requires numeric values only"



  # Handle multiplication of numbers

  elif operation == "multiply":
    try:
      result = 1
      for num in args:
        result *= num
      if "round_to" in options:
        result = round(result, options["round_to"])
      return result
    except TypeError:
      return "Error: 'multiply' operation requires numeric values only"


  # Handle elif operation
  elif operation == "concat":
    try:
      result = "".join(str(item) for item in args)
      if options.get("uppercase", False):
        result = result.upper()
      return result
    except Exception as e:
      return f"Error during concatenation: {e}"

  else:
    return "Error: Unsupported operation. Use 'add', 'multiply' or 'concat' "


print(flexible_function("add", 3, 5, 2.4))
# ➜ 10.4

print(flexible_function("add", 3.14159, 2.71828, round_to=2))
# ➜ 5.86

print(flexible_function("multiply", 2, 3, 4))
# ➜ 24

print(flexible_function("concat", "data", "science", "rocks", uppercase=True))
# ➜ "DATASCIENCEROCKS"

print(flexible_function("concat", "Hello", " ", "World"))
# ➜ "Hello World"

print(flexible_function("divide", 2, 3))
# ➜ "Error: Unsupported operation. Use 'add', 'multiply', or 'concat'."

```

```python
# Unpacking arguments


def func(a,b,c):
  print(a)
  print(b)
  print(c)


func(*[1,2,3])


def dict_func(name, age, city):
  print(name)
  print(age)
  print(city)


dict_func(**{'name': "Alice", 'age': "in the", 'city':"Borderland"})



```

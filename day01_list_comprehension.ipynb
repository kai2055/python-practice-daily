```python
# Creare a list of squares for numbers 1 to 10

square_numbers = [x * x for x in range(1,11)]
print(square_numbers)
```

```python
# Create a list of even numbers from o to 20

even_numbers = [n for n in range(0,21) if n % 2 == 0]
print(even_numbers)
```

```python
# Convert all strings in this list to uppercase

words = ["hello", "world", "python", "rocks"]

upper_converted = [word.upper() for word in words]
print(upper_converted)
```

```python
# Get the length of each word in the list

fruits = ["apple", "banana", "cherry", "data"]

fruits_length = [len(fruit) for fruit in fruits]
print(fruits_length)
```

```python
# Get all numbers divisible by 3 or 5 from 1 to 50

divisible_numbers = [num for num in range(1,51) if num % 3 == 0 or num % 5 == 0]
print(divisible_numbers)
```

```python
# Extract first character of each word

animals = ["cat", "dog", "elephant", "fox"]

first_character = [word[0] for word in animals]
print(first_character)
```

```python
# Create list of tuples (number, square) for numbers 1-5

square_number_tuple = [(num, num*num) for num in range(1,6)]
print(square_number_tuple)
```

```python
# Filter out negative numbers and square the rest

numbers = [3, -2, 7, -1, 5, -4, 8]

filtered_square_numbers = [num*num for num in numbers if num > 0]
print(filtered_square_numbers)
```

```python
# Flatten a 2D list

matrix = [[1,2,3], [4,5,6], [7,8,9]]

flattened_matrix = []

for i in range(len(matrix)):
  for j in range(len(matrix[i])):
    flattened_matrix.append(matrix[i][j])
print(flattened_matrix)


flattened_matrix = [matrix[i][j] for i in range(len(matrix)) for j in range(len(matrix[i]))]
print(flattened_matrix)
```

```python
# Create list of all vowels found in a sentence

sentence = "The quick brown fox jumps over the lazy dog"

vowels_list = [vowel for vowel in sentence if vowel in ['a', 'e', 'i', 'o', 'u']]
print(vowels_list)
```

```python
# Get all prime numbers from 2 to 50

import math
prime_num = [num for num in range(2, 50) if not any(num % i == 0 for i in range(2, int(math.sqrt(num))+1))]
print(prime_num)

```

```python
prime_num = [num for num in range(2, 50)
             if all(num % i != 0 for i in range(2,num)) ]
print(prime_num)
```

```python
# Create a dictionary from two lists using dict comprehension

keys = ["name", "age", "city"]
values = ["Alice", 25, "NYC"]


dict_comprehension = {k:v for k,v in zip(keys, values)}
print(dict_comprehension)

person = dict(zip(keys, values))
print(person)
```

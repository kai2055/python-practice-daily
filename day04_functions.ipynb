```python
# Write a function to calculate Fibonacci number at position n
# Example: fibonacci(6) -> 8 (Sequence: 0,1,1,2,3,5,8)


def fibonacci(term):
  fib_series = [0,1]

  if term == 0:
    return 0
  elif term == 1:
    return 1
  elif term == 2:
    return 0 + 1
  else:
    for i in range(2, term + 1):
      next_fib = fib_series[i-1] + fib_series[i-2]
      fib_series.append(next_fib)

    return fib_series[term]



print(fibonacci(6))



```
